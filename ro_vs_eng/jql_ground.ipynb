```python
import json
from sklearn.metrics import root_mean_squared_error, f1_score, accuracy_score
from scipy.stats import spearmanr
from statistics import mean
```

```python
with open('jql_ground.json', 'r') as file:
    data = json.load(file)

human, test, jql = data
```

```python
for dataset in [human, test, jql]:
    labels = dataset['int_scores']
    ours = dataset['reg_preds']
    jql_gemma = dataset['jql_preds']['Edu-JQL-Gemma-SF']
    jql_mistral = dataset['jql_preds']['Edu-JQL-Mistral-SF']
    jql_lamma = dataset['jql_preds']['Edu-JQL-Llama-SF']
    jql_average = [mean(x) for x in zip(jql_gemma, jql_mistral, jql_lamma)]

    print(f"Dataset: {dataset['dataset']}")
    print(f"RMSE (ours): {root_mean_squared_error(labels, ours)}")
    print(f"RMSE (jql_gemma): {root_mean_squared_error(labels, jql_gemma)}")
    print(f"RMSE (jql_mistral): {root_mean_squared_error(labels, jql_mistral)}")
    print(f"RMSE (jql_lamma): {root_mean_squared_error(labels, jql_lamma)}")
    print(f"RMSE (jql_average): {root_mean_squared_error(labels, jql_average)}")
    print()
    print(f"Spearman (ours): {spearmanr(labels, ours).correlation}")
    print(f"Spearman (jql_gemma): {spearmanr(labels, jql_gemma).correlation}")
    print(f"Spearman (jql_mistral): {spearmanr(labels, jql_mistral).correlation}")
    print(f"Spearman (jql_lamma): {spearmanr(labels, jql_lamma).correlation}")
    print(f"Spearman (jql_average): {spearmanr(labels, jql_average).correlation}")
    print()
    print("-" * 40, "\n")
```
